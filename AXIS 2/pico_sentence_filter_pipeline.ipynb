# Final Decomposed PICO Extraction with Bio_ClinicalBERT


## Task

The earlier version of the decomposed pipeline had a strong sentence stage but a weak token stage.

The main hypothesis in this notebook is:

1. sentence filtering can work reasonably well
2. token extraction suffers if we use only coarse BIO labels
3. token extraction should improve when we use the **fine-grained hierarchical labels** already available in EBM-NLP:
   - Participants: Age, Sex, Sample size, Condition
   - Interventions: Surgical, Physical, Drug, Educational, Psychological, Other, Control
   - Outcomes: Physical, Pain, Mortality, Adverse effects, Mental, Other

So this notebook explicitly uses those subtype labels and tracks whether the bottleneck is:

- the sentence stage
- the token stage
- or the interaction between both stages

## 5. Inspecting Label Alignment



```python
def preview_alignment(entity_type: str, split: str = "train", max_rows: int = 40):
    docs = cte.load_dataset(DATA_DIR, entity_type, split)
    for doc in docs:
        if any(label_id != 0 for label_id in doc.label_ids):
            rows = []
            for idx, (tok, label_id, tag) in enumerate(zip(doc.tokens, doc.label_ids, doc.tags)):
                if label_id != 0:
                    start = max(0, idx - 5)
                    end = min(len(doc.tokens), idx + max_rows)
                    for j in range(start, end):
                        rows.append(
                            {
                                "doc_id": doc.doc_id,
                                "index": j,
                                "token": doc.tokens[j],
                                "label_id": doc.label_ids[j],
                                "typed_bio_tag": doc.tags[j],
                            }
                        )
                    return pd.DataFrame(rows).drop_duplicates(subset=["index"]).reset_index(drop=True)
    return pd.DataFrame()


print("Participants example")
display(preview_alignment("participants"))

print("Interventions example")
display(preview_alignment("interventions"))

print("Outcomes example")
display(preview_alignment("outcomes"))
```

```python
sentence_summary_rows = []
for entity_type in dp.ENTITY_TYPES:
    train_docs = cte.load_dataset(DATA_DIR, entity_type, "train")
    test_docs = cte.load_dataset(DATA_DIR, entity_type, "test")
    train_sentences = dp.docs_to_binary_sentence_examples(train_docs, entity_type)
    test_sentences = dp.docs_to_binary_sentence_examples(test_docs, entity_type)
    sentence_summary_rows.append(
        {
            "entity_type": entity_type,
            "train_sentences": len(train_sentences),
            "test_sentences": len(test_sentences),
            "train_positive": sum(x.is_positive for x in train_sentences),
            "train_negative": len(train_sentences) - sum(x.is_positive for x in train_sentences),
            "test_positive": sum(x.is_positive for x in test_sentences),
            "test_negative": len(test_sentences) - sum(x.is_positive for x in test_sentences),
        }
    )

sentence_summary_df = pd.DataFrame(sentence_summary_rows)
display(sentence_summary_df)
```

## 13. Token-Level Results

The token stage is reported in two ways:

1. **pipeline token metrics**: the real decomposed system
2. **oracle token metrics**: token extraction given the gold sentence filter

The difference between them is crucial.

If oracle token F1 is much higher than pipeline token F1, then sentence filtering is throwing away useful signal.

If both are low, then the token extractor itself is still weak.

## 21. Comparator Evaluation Against Gold Labels

We evaluate Comparator against the gold `CONTROL` subtype inside the intervention labels.

Gold Comparator:
- intervention tokens tagged as `B-CONTROL` or `I-CONTROL`

Predicted Comparator:
- predicted intervention tokens tagged as `B-CONTROL` or `I-CONTROL`

This gives direct token-level and span-level Comparator metrics.


```python
def comparator_binary_tags(tags):
    binary = []
    prev_is_control = False

    for tag in tags:
        is_control = tag.endswith("CONTROL")
        if not is_control:
            binary.append("O")
            prev_is_control = False
        else:
            binary.append("I" if prev_is_control else "B")
            prev_is_control = True

    return binary


def binary_spans(tags):
    spans = []
    start = None

    for idx, tag in enumerate(tags):
        if tag == "B":
            if start is not None:
                spans.append((start, idx - 1))
            start = idx
        elif tag == "O":
            if start is not None:
                spans.append((start, idx - 1))
                start = None
        elif tag == "I" and start is None:
            start = idx

    if start is not None:
        spans.append((start, len(tags) - 1))

    return spans


def evaluate_comparator_from_artifact(artifact):
    tp = fp = fn = 0
    token_correct = 0
    token_total = 0

    for ex in artifact.test_sentences:
        key = (ex.doc_id, ex.sent_idx)

        gold_tags = comparator_binary_tags(ex.tags)
        pred_tags = comparator_binary_tags(artifact.token_sentence_pred_map[key])

        gold_spans = set(binary_spans(gold_tags))
        pred_spans = set(binary_spans(pred_tags))

        tp += len(gold_spans & pred_spans)
        fp += len(pred_spans - gold_spans)
        fn += len(gold_spans - pred_spans)

        token_correct += sum(g == p for g, p in zip(gold_tags, pred_tags))
        token_total += len(gold_tags)

    precision = tp / (tp + fp) if tp + fp else 0.0
    recall = tp / (tp + fn) if tp + fn else 0.0
    f1 = 2 * precision * recall / (precision + recall) if precision + recall else 0.0
    token_accuracy = token_correct / token_total if token_total else 0.0
    coverage = recall  # span-level gold recovery rate

    return {
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "token_accuracy": token_accuracy,
        "coverage_pct": coverage * 100,
        "tp": tp,
        "fp": fp,
        "fn": fn,
    }


comparator_metrics_rows = []
for strategy in ["zero-shot", "few-shot"]:
    metrics = evaluate_comparator_from_artifact(artifacts[("interventions", strategy)])
    comparator_metrics_rows.append(
        {
            "entity_type": "comparator",
            "strategy": strategy,
            "precision": metrics["precision"],
            "recall": metrics["recall"],
            "f1": metrics["f1"],
            "token_accuracy": metrics["token_accuracy"],
            "coverage_pct": metrics["coverage_pct"],
            "tp": metrics["tp"],
            "fp": metrics["fp"],
            "fn": metrics["fn"],
        }
    )

comparator_metrics_df = pd.DataFrame(comparator_metrics_rows)
display(comparator_metrics_df)
```

```python
def sentence_has_comparator(tags):
    return int(any(tag.endswith("CONTROL") for tag in tags))


def evaluate_comparator_sentence_level(artifact):
    gold = []
    pred = []

    for ex in artifact.test_sentences:
        key = (ex.doc_id, ex.sent_idx)

        gold_value = sentence_has_comparator(ex.tags)
        pred_value = sentence_has_comparator(artifact.token_sentence_pred_map[key])

        gold.append(gold_value)
        pred.append(pred_value)

    tp = sum(g == 1 and p == 1 for g, p in zip(gold, pred))
    fp = sum(g == 0 and p == 1 for g, p in zip(gold, pred))
    fn = sum(g == 1 and p == 0 for g, p in zip(gold, pred))
    tn = sum(g == 0 and p == 0 for g, p in zip(gold, pred))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0
    accuracy = (tp + tn) / len(gold) if gold else 0.0
    keep_rate = sum(pred) / len(pred) if pred else 0.0

    return {
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "accuracy": accuracy,
        "keep_rate": keep_rate,
        "tp": tp,
        "fp": fp,
        "fn": fn,
        "tn": tn,
    }


comparator_sentence_rows = []
for strategy in ["zero-shot", "few-shot"]:
    metrics = evaluate_comparator_sentence_level(artifacts[("interventions", strategy)])
    comparator_sentence_rows.append(
        {
            "entity_type": "comparator",
            "strategy": strategy,
            "sentence_accuracy": metrics["accuracy"],
            "sentence_precision": metrics["precision"],
            "sentence_recall": metrics["recall"],
            "sentence_f1": metrics["f1"],
            "sentence_keep_rate": metrics["keep_rate"],
            "tp": metrics["tp"],
            "fp": metrics["fp"],
            "fn": metrics["fn"],
            "tn": metrics["tn"],
        }
    )

comparator_sentence_metrics_df = pd.DataFrame(comparator_sentence_rows)
display(comparator_sentence_metrics_df)
```

```python
# Main entity sentence-stage metrics
main_sentence_df = summary_df[
    [
        "entity_type",
        "strategy",
        "sentence_accuracy",
        "sentence_precision",
        "sentence_recall",
        "sentence_f1",
        "sentence_keep_rate",
    ]
].copy()

# Main entity token-stage metrics
main_token_df = summary_df[
    [
        "entity_type",
        "strategy",
        "token_pipeline_precision",
        "token_pipeline_recall",
        "token_pipeline_f1",
        "token_pipeline_token_accuracy",
    ]
].copy().rename(
    columns={
        "token_pipeline_precision": "token_precision",
        "token_pipeline_recall": "token_recall",
        "token_pipeline_f1": "token_f1",
        "token_pipeline_token_accuracy": "token_accuracy",
    }
)

# Main entity coverage as percentage
main_coverage_df = main_token_df[["entity_type", "strategy", "token_recall"]].copy()
main_coverage_df["coverage_pct"] = main_coverage_df["token_recall"] * 100
main_coverage_df = main_coverage_df.drop(columns=["token_recall"])

# Merge main entity results
main_results_df = (
    main_sentence_df
    .merge(main_token_df, on=["entity_type", "strategy"], how="inner")
    .merge(main_coverage_df, on=["entity_type", "strategy"], how="inner")
)

# Comparator results
comparator_results_df = (
    comparator_sentence_metrics_df[
        [
            "entity_type",
            "strategy",
            "sentence_accuracy",
            "sentence_precision",
            "sentence_recall",
            "sentence_f1",
            "sentence_keep_rate",
        ]
    ]
    .merge(
        comparator_metrics_df[
            [
                "entity_type",
                "strategy",
                "precision",
                "recall",
                "f1",
                "token_accuracy",
                "coverage_pct",
            ]
        ].rename(
            columns={
                "precision": "token_precision",
                "recall": "token_recall",
                "f1": "token_f1",
            }
        ),
        on=["entity_type", "strategy"],
        how="inner",
    )
)

# Final combined table
final_results_df = pd.concat(
    [main_results_df, comparator_results_df],
    ignore_index=True,
)

entity_order = ["participants", "interventions", "outcomes", "comparator"]
strategy_order = ["zero-shot", "few-shot"]

final_results_df["entity_type"] = pd.Categorical(
    final_results_df["entity_type"],
    categories=entity_order,
    ordered=True,
)
final_results_df["strategy"] = pd.Categorical(
    final_results_df["strategy"],
    categories=strategy_order,
    ordered=True,
)

final_results_df = final_results_df.sort_values(
    ["entity_type", "strategy"]
).reset_index(drop=True)

display(final_results_df)
```
